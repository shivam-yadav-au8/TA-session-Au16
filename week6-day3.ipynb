```python
def fun(s,i,l,c):
    if(i==l):
        return c
    if(s[i]=='a,A' or s[i]=='e' or s[i]=='o' or s[i]=='i' or s[i]=='u' ):
        c+=1
    return fun(s,i+1,l,c)
s=input()
print(fun(s,i=0,l=len(s),c=0))
```

```python
def fun(c,a,b,s,inp):
    if(c>=inp):
        return s 
    s=a+b
    return fun(c+1,b,s,s,inp)
inp=int(input())
if(inp==1):
    print(0)
if(inp==2):
    print(1)
else:
    c=2
    a=0
    b=1
    s=0
    print(fun(c,a,b,s,inp))
```

```python
def fun(s,c):
    if(c==len(s)-1):
        return c+1
    elif(c>len(s)-1):
        return 0
    return fun(s,c+1)
s="dcdc"
c=0
print(fun(s,c))
```

```python
def fun(c,inp,s):
    if(c==inp):
        return s+c
    s+=c
    return fun(c+1,inp,s)
inp=int(input())
c=1
s=0
print(fun(c,inp,s))
```

```python
def fun(base,exp,a):
    if(exp==0):
        return a
    return fun(base,exp-1,a*base)
base=int(input())
exp=int(input())
a=1
print(fun(base,exp,a))
```
