```python
class Time:
    def __init__(self,hours,minutes):
        self.hours=hours
        self.minutes=minutes
    def addTime(self,time1,time2):
        print(time1.hours+time2.hours,time1.minutes+time2.minutes)
obj=Time(5,40)
obj1=Time(4,14)
obj2=Time(8,54)
obj3=Time(9,23)
obj.addTime(obj1,obj2)
obj.addTime(obj3,obj2)
```
